```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
# DivisiveClusterive is not implemented in libraries usually 
from scipy.cluster.hierarchy import dendrogram, linkage

# Sample dataset
X = np.array([
    [1,2],
    [2,3],
    [3,4],
    [8,7],
    [9,8],
    [10,9]
])

# Step 1: Create linkage matrix for dendrogram
Z = linkage(X, method='ward')

# Step 2: Plot dendrogram
plt.figure(figsize=(6,4))
dendrogram(Z)
plt.title("Dendrogram for Hierarchical Clustering")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()

# Step 3: Apply Agglomerative Hierarchical Clustering
model = AgglomerativeClustering(n_clusters=2)
labels = model.fit_predict(X)

# Step 4: Print cluster labels
print("Cluster Labels:", labels)

# Step 5: Visualize clustered data
plt.scatter(X[:,0], X[:,1], c=labels)
plt.title("Hierarchical Clustering Result")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
