```python
from sklearn.cluster import DBSCAN
import numpy as np

# Sample data
X = np.array([
    [1, 2],
    [2, 2],
    [2, 3],
    [8, 7],
    [8, 8],
    [25, 80]
])

# Create DBSCAN model
model = DBSCAN(eps=1.5, min_samples=2)

# Fit the model
labels = model.fit_predict(X)

print("Cluster labels:", labels)
```
