```python
# without K fold 
# 1️⃣ Import libraries
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score

# 2️⃣ Sample dataset (small demo dataset)
emails = [
    "Congratulations you won a free lottery",
    "Win money now click here",
    "Limited offer claim your prize",
    "Hey are we meeting tomorrow?",
    "Let's have lunch today",
    "Can you send the project files?"
]

labels = [
    "spam",
    "spam",
    "spam",
    "ham",
    "ham",
    "ham"
]

# 3️⃣ Convert text to TF-IDF vectors
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(emails)

# 4️⃣ Split data
X_train, X_test, y_train, y_test = train_test_split(
    X, labels, test_size=0.3, random_state=42
)

# 5️⃣ Train SVM (Linear Kernel)
model = SVC(kernel="linear")
model.fit(X_train, y_train)

# 6️⃣ Make predictions
predictions = model.predict(X_test)

# 7️⃣ Evaluate
print("Accuracy:", accuracy_score(y_test, predictions))
print("\nClassification Report:\n")
print(classification_report(y_test, predictions))
```

```python
# with k fold and pipeline
# 1️⃣ Imports
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from sklearn.model_selection import KFold, cross_val_score

# 2️⃣ Sample dataset
emails = [
    "Congratulations you won a free lottery",
    "Win money now click here",
    "Limited offer claim your prize",
    "Hey are we meeting tomorrow?",
    "Let's have lunch today",
    "Can you send the project files?",
    "Exclusive deal just for you",
    "Meeting postponed to next week",
    "Free vacation offer click now",
    "Submit your assignment by tonight"
]

labels = [
    "spam",
    "spam",
    "spam",
    "ham",
    "ham",
    "ham",
    "spam",
    "ham",
    "spam",
    "ham"
]

# 3️⃣ Create Pipeline (VERY IMPORTANT)
model = Pipeline([
    ("tfidf", TfidfVectorizer()),
    ("svm", SVC(kernel="linear", C=1))
])

# 4️⃣ Define K-Fold
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# 5️⃣ Perform Cross Validation
scores = cross_val_score(model, emails, labels, cv=kf, scoring="accuracy")

# 6️⃣ Results
print("Accuracy for each fold:", scores)
print("Mean Accuracy:", scores.mean())
```
