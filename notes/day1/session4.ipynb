```python
# Gaussian Naive Bayes

import numpy as np
from sklearn.naive_bayes import GaussianNB

# Features: [height (cm), weight (kg), foot size (cm)]
X = np.array([
    [180, 80, 26],   # Male
    [170, 75, 25],   # Male
    [175, 78, 27],   # Male
    [160, 55, 22],   # Female
    [165, 60, 23],   # Female
    [155, 50, 21]    # Female
])

# Labels
y = np.array(['Male', 'Male', 'Male', 'Female', 'Female', 'Female'])

# Create model
model = GaussianNB()

# Train model
model.fit(X, y)

# Predict new person
new_person = np.array([[172, 70, 24]])
prediction = model.predict(new_person)

print("Predicted Gender:", prediction[0])
```

```python
# Multinomial Naive Bayes

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

# Training emails
emails = [
    "Win money now",
    "Claim your free prize",
    "Meeting tomorrow at office",
    "Project discussion schedule"
]

# Labels
labels = ["Spam", "Spam", "Not Spam", "Not Spam"]

# Convert text to word count vectors
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(emails)

# Create model
model = MultinomialNB()

# Train model
model.fit(X, labels)

# Predict new email
new_email = ["Free money prize"]
new_X = vectorizer.transform(new_email)

prediction = model.predict(new_X)

print("Prediction:", prediction[0])
```
